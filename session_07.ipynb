```python
import pandas as pd
```

```python
data = pd.read_csv('Coffee_company.csv')
```

```python
# removing any leading/trailing spaces from column names
data.columns = data.columns.str.strip()
```

```python
data.head() # displaying the first few rows of the dataset
```

```python
# 1. Round up 'Units Sold' to the nearest integer
data['Units Sold'] = data['Units Sold'].apply(lambda x: int(x + (x % 1 > 0)))
data.head()

# import math
# data['Units Sold'] = data['Units Sold'].apply(lambda x: math.ceil(x))
# data.head()
```

```python
print(data.columns) #to check the exact name of each column
```

```python
# 2. Rename the column 'Month Name' to 'Month'
data.rename(columns={'Month Name': 'Month'}, inplace=True)
data.head()
```

```python
# Remove unnecessary characters and convert to numeric
data['Sales'] = pd.to_numeric(data['Sales'].str.replace('$', '', regex=False).str.replace(',', '', regex=False), errors='coerce')
data['Sales'] = data['Sales'].astype(float)

# Display the updated DataFrame
print(data.head())
```

```python
# Calculate total sales
total_sales = data['Sales'].sum()
print("Total_Sales: $", total_sales)
```
